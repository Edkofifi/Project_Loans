## Investigation Overview


> The focus of this investigation is to determine the main attributes of a loan data that influence the returns that lenders get on their investment (monies lent to borrowers). Three main attributes - borrower's annual percentage rate, loan ratings and the estimated loss of the loans will be investigated.

## Dataset Overview

> A loan data containing 113,937 rows was used for the investigation. The initial data contains 81 attributes. Only 23 of those attributes were needed for this analysis hence the rest were not dropped. Some of the important attributes in the data include the lenders' yield,loan ratings,  loan status, estimated loss, borrowers' annual percentage rate.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import pickle
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loans = pd.read_pickle('clean_loans.pickle')
```

## Distribution of Lenders' Yield

> When plotted on an interval of 2%, we noticed one major return peaks in the frequency - 15% returns to the lenders. Generally, the distribution is normal.


```python
# Plot the lender's yield on a standard scale
plt.figure(figsize=[8,6])
binsize = 0.02
bins = np.arange(0, loans['LenderYield'].max()+binsize, binsize)
plt.hist(data = loans, x = 'LenderYield', bins = bins )
plt.xlabel('Lenders\'  yield')
plt.ylabel('Count')
plt.title('Distribution of Lenders\' Yield')
plt.show()
```

## How the Original Loan Amount is Distributed

> The distribution shows peak number of loans originally borrowed at $\$$4,000.00, $\$$5,000.00,  $\$$10,000.00, $\$$15,000.00, $\$$20,000.00, $\$$20,000.00, and $\$$35,000.00. These may represent standard borrowed amounts in the banks.

```python
plt.figure(figsize=[8,6])
# plotting original loan amount on smaller bin size
binsize = 500
bins = np.arange(0, loans['LoanOriginalAmount'].max()+binsize, binsize)
plt.hist(data = loans, x = 'LoanOriginalAmount', bins=bins )
plt.xlabel('Original Loan Amount (in $)')
plt.ylabel('Count')
plt.title('Distribution of the Loan Original Amount')
plt.show()
```

## Distribution of the Borrowers' Annual Percentage Cost

>The cost of borrowing peak 20% while the plot is fairly normal.

```python
plt.figure(figsize=[8,6])
binsize = 0.03
bins = np.arange(0, loans['BorrowerAPR'].max()+binsize, binsize)
plt.hist(data = loans, x = 'BorrowerAPR', bins = bins )
plt.xlabel('Borrowers\' Annual Percentage Rate')
plt.ylabel('Count')
plt.title('Distribution of the cost of borrowing')
plt.show()
```

## Lenders' Yield Vrs Borrowers' Annual Percentage Rate
> On the logarithmic scale, the lenders' yield and the borrower annual percentage rate indicates a strong postive correlation (0.99). As borrower's rate increases, lenders' yield increases. The chart also indicates a slightly reduced rate of increment after it reaches 20% annual percentage rate. At that point the lenders' yield also begin to increase gradually from 15%. High 
borrowers' annual percentage rates depict high risky investment hence could explain increasing lenders' yield with increasing borrowers' rates.

```python
# We plot the LenderYield and the BorrowerAPR on a log scale
plt.figure(figsize = [8, 6])
plt.scatter(data = loans, x = 'LenderYield', y = 'BorrowerAPR', alpha = 0.5/10)
plt.xlabel('Borrower Annual Percentage Cost')
plt.yscale('log')
plt.xlim([0.03, 0.35])
plt.yticks([10e-2, 10e-1], [0.1, 0.2])
plt.ylabel('Lender Yield (%)')
plt.title('Relationship between Lenders\' Yield and Borrowing Cost')
plt.show()
```

## Lenders' Yield Vrs Estimated Losss
> Though not strong as the borrowers' rate, the estimated loss also signifies an important relationship with the lenders' yield. A few points reflect no relationship but generally the impact of the estimated is strong (0.83). The estimated loss predicted on the loan investment will generally raise lenders' expectations of good returns to enable them invest. We see that higher estimated loss is associated with higher lender yield as shown in the graph.

```python
# We plot the LenderYield and the Estimated loss on a log scale
plt.figure(figsize = [8, 6])
plt.scatter(data = loans, x = 'LenderYield', y = 'EstimatedLoss', alpha = 0.1/10)
plt.xlabel('Estimated Loss on a loan')
plt.yscale('log')
plt.xlim([0.03, 0.35])
plt.yticks([10e-2, 10e-1], [0.1, 0.2])
plt.ylabel('Lender Yield (%)')
plt.show()
```

## Lenders' Yield by Ratings and Home ownership
> In plotting the lenders' yield and the prosper rating with whether a borrower is a home owner or not, the plot indicates that the ordered prosper ratings has a strong relationship with the yield that lenders get. Loans that are rated lowest have a highest yield, up to 31% returns to investors. Highly rated loans turns to have lower returns. 

```python
# plot lenders' yield, prosper ratings and home ownership using box plot
plt.figure(figsize=[10,6])
ax = sb.boxplot(data = loans, x = 'ProsperRating (Alpha)', y = 'LenderYield', hue = 'IsBorrowerHomeowner')
ax.legend(title = 'Is borrower a homeowner')
ax.set_xlabel("Loan Ratings (Alpha) ")
ax.set_ylabel("Lenders' Yield ")
ax.set_title("Lenders' Yield Vrs Prosper Ratings and Home ownership");
```
